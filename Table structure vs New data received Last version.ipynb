```python
import pandas as pd
import os
import glob
import numpy as np
from decimal import *

path = 'C:\\Users\\AROLMCAC\\Desktop\\Test python\\UPLOAD_DATA'

files = os.listdir(path)

for file in files:
    extension = os.path.splitext(file)[1]

print(extension)

result = [i for i in glob.glob(path + "\*.csv")+ glob.glob(path + "\*.xlsx")]

if extension == '.xlsx':
    df = pd.DataFrame()

    for file in result:
        temp = pd.read_excel(file)
        df = pd.concat([df, temp], ignore_index=True, sort=False, axis =0)  
    for column in df.columns:
        print(column,"|", df[column].dtype.name,"|",df[column].astype(str).str.len().max())
    
    TEST1 = []
    for column in df.columns:
        TEST1.append(
        {   'TECHNICAL_NAME': column,
            'Column Datatype': df[column].dtype.name,
            'File length Column': df[column].astype(str).str.len().max()})
        
    MT= pd.DataFrame(TEST1) #create dataframe 

    System_name = 'CEGID' #-------------------> fill with system name 

    System = (r'C:\Users\AROLMCAC\Desktop\Test python\Systems\C11_{}_Table_SQL.xlsx'.format(System_name)) #-------------> required to change table structure

    test1 = pd.read_excel(System)
 
    DF_MT=pd.merge(MT,test1, on='TECHNICAL_NAME')

    DF_MT['LENGTH'] = DF_MT['LENGTH'].astype(int)

    DF_MT['Status'] = np.where(DF_MT['File length Column'] > DF_MT['LENGTH'], "NOT OK","OK" )

    dflast = DF_MT.groupby( ['TECHNICAL_NAME'])['Column Datatype','TYPE', 'File length Column', 'LENGTH', 'Status' ].max()
    
    dflast.head()

    dflast.to_excel("C:\\Users\\AROLMCAC\\Desktop\\test2.xlsx")
    
    pass
    
elif extension == '.csv':
    df = pd.DataFrame()

    for file in result:
        temp = pd.read_csv(file,sep=',', decimal = '.', na_filter=False, low_memory = False, encoding= 'utf8', dtype=str)
        df = pd.concat([temp, df], ignore_index=True, sort=False, axis =0)

    df.fillna(0,inplace= True)

    for column in df.columns:
        print(column,"|", df[column].dtype.name,"|",df[column].astype(str).str.len().max())
    
    TEST1 = []
    for column in df.columns:
        TEST1.append(
        {
            'TECHNICAL_NAME': column,
            'Column Datatype': df[column].dtype.name,
            'File length Column': df[column].astype(str).str.len().max()})
    
    MT= pd.DataFrame(TEST1) #create dataframe 

    System_name = 'ECC' #-------------------> fill with system name 

    System = (r'C:\Users\AROLMCAC\Desktop\Test python\Systems\P11_{}_Table_SQL.xlsx'.format(System_name)) #-------------> required to change table structure

    test1 = pd.read_excel(System)
 
    DF_MT=pd.merge(MT,test1, on='TECHNICAL_NAME')

    DF_MT['LENGTH'] = DF_MT['LENGTH'].astype(int)

    DF_MT['Status'] = np.where(DF_MT['File length Column'] > DF_MT['LENGTH'], "NOT OK","OK" )

    dflast = DF_MT.groupby( ['TECHNICAL_NAME'])['Column Datatype','TYPE', 'File length Column', 'LENGTH', 'Status' ].max()
    
    dflast.head()

    dflast.to_excel("C:\\Users\\AROLMCAC\\Desktop\\test2.xlsx")
    
    pass
else:
    print('No extension')
```
