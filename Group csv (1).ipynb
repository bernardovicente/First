```python
import pandas as pd
import glob
import numpy as np
from decimal import *

path = "C:\\Users\\AROLMCAC\\Downloads\\drive-download-20221115T160904Z-001"
  
filenames = glob.glob(path + "\*.csv")
print('File names:', filenames)
  
df = pd.DataFrame()

for file in filenames:
  
    temp = pd.read_csv(file, sep =';', decimal = '.', na_filter=False, low_memory = False)
    df = pd.concat([df, temp], ignore_index=True, sort=False)
  

df.drop('Unnamed: 0', axis=1, inplace=True)    

df.to_csv(r'C:\\Users\\AROLMCAC\\Desktop\\file2.csv', header=False, index=False)
#with pd.ExcelWriter(r'C:\\Users\\AROLMCAC\\Desktop\\File.xlsx') as writer:
    #df.to_excel(writer, sheet_name='Final',index = False)

df.head()
```
